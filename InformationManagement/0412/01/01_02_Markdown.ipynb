# 第1回 その2: Markdown記法による文章の書き方  
ここでは，Jupyter Notebook 上で文章を書くための基本的な方法を説明します。  
Jupyter Notebookにおける文章は，Markdown（マークダウン）という記法を用いて記述します。

## ステップ1: 改行について
初めてMarkdownを使った時に違和感を覚えるのは「改行」の方法だと思います。  
通常のテキストエディタと同じように二つの文を改行して書いても，Jupyter Notebook上では改行してくれません。  
```
(テキストエディタ上での表示)
1行目
2行目
```
↓
```
(Jupyter Notebook 上での表示)
1行目2行目
```
Markdown記法で改行を行う場合は，以下3つの方法があります。  
1. 改行したい文の末尾に半角スペースを二つ入れる（全角スペースはダメ）。  
  1行目 <font color="Red">_ _</font>  
  2行目
2. 改行したい文の下に1行空白の行を入れる。  
  1行目
  <br><br>
  2行目  
3. 改行タグ\<br\>を入れる。  
  1行目\<br\>2行目

どれを使っても良いですが，方法2を使った場合のみ，文章間隔が若干広くなります。  
***
半角スペースを使った場合1行目  
半角スペースを使った場合2行目  
***
空白行を入れた場合1行目

空白行を入れた場合2行目  
***
改行タグ\<br\>を使った場合1行目<br>
改行タグ\<br\>を使った場合2行目
***
好みのものを使用してください。  

## ステップ2: 文字の修飾  
文字の大きさを変えたり色を付けたり太字にしたりできます。  
見出し（このページの第1回やステップ2といった部分）は文頭に # を付けることで定義できます。  
```
# 見出し1
## 見出し2
### 見出し3
#### 見出し4
```
# 見出し1
## 見出し2
### 見出し3
#### 見出し4

***
色を付ける場合は以下のようにします。  
```
文章中の<font color="Red">ここの部分</font>が赤色になります。
```
文章中の<font color="Red">ここの部分</font>が赤色になります。

***
太字にする場合は以下のようにします。
```
文章中の**ここの部分**が太字になります。
```
文章中の**ここの部分**が太字になります。



## ステップ3: 箇条書き
箇条書きは以下のように書きます。  
```
* 項目1
  * 項目1-1
  * 項目1-2
    * 項目1-2-1
* 項目2
* 項目3
# ここに空白行を入れて箇条書きを終える
```
* 項目1
  * 項目1-1
  * 項目1-2
    * 項目1-2-1
* 項目2
* 項目3

箇条書きを終えたい場合は空白行をいれる方法で改行をします。  

 ## ステップ4: おわりに  
他にも様々な書式で文章を書くことができます。  
「Markdown記法」で検索すると色々なテクニックを知ることができるはずです。
